```python
songsFrequency=pd.read_csv("https://raw.githubusercontent.com/BollywoodData/Bollywood-Data/master/wikipedia-data/songsFrequency.csv")
```

```python
#Putting it in a DataFrame
```

```python
dfsongs = pd.DataFrame(songsFrequency)
dfsongs
```

```python
#Cleaning the attributes by removing the space between the attributes and replacing with "_"
```

```python
#dfsongs.rename(columns = {"YEAR":"year"},inplace=True)
dfsongs.columns = dfsongs.columns.str.replace(' ','_')
dfsongs
```

```python
#Getting the values of the 20-30th row in the dataframe for visualization
```

```python
Head = dfsongs.iloc[20:30,:]
Head
```

```python
#Extracting the cells which are necessary for visualization
```

```python
Select = Head[["_YEAR","_MALE_SONG_COUNT","_FEMALE_SONG_COUNT"]]
#Select["Song_count"] = Head_song_count
Select
```

```python
#1.Getting the current axis using gca()
#2.Using plt(imported using matplotlib) to create scatter plot for year vs female_song_count
#3.Using plt(imported using matplotlib) to create scatter plot for year vs male_song_count
#4.Plotting both the lines in same graph
#5.Showing the output using plt.show()
```

```python
ax =plt.gca()
plt.scatter(x=Select._YEAR,y=Select._FEMALE_SONG_COUNT)
plt.scatter(x=Select._YEAR,y=Select._MALE_SONG_COUNT)
Select.plot(kind="line",x="_YEAR",y="_FEMALE_SONG_COUNT",ax=ax)

Select.plot(kind="line",x="_YEAR",y="_MALE_SONG_COUNT",ax=ax)
plt.show()
```
